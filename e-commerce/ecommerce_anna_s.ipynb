# Проект "E-commerce"

<b>Цель:</b>
Сегментация покупателей по профилю потребления.

<b>Задачи:</b>
* Провести исследовательский анализ данных.
* Сегментировать покупателей согласно истории их покупок. Сформировать сегменты на основе того, что покупают пользователи: провести анализ их покупок и сегментировать товары по принадлежности к товарной категории.
* Сформулировать и проверить статистические гипотезы.

<b>Описание данных:</b>

Датасет описывает транзакции интернет-магазина товаров для дома и быта «Пока все ещё тут».

Колонки в ecommerce_dataset.csv :
* date — дата заказа;
* customer_id — идентификатор покупателя;
* order_id — идентификатор заказа;
* product — наименование товара;
* quantity — количество товара в заказе;
* price — цена товара.

* Предобработка данных:
    * Удаление пропусков.
    * Удаление явных и неявных дубликатов.
    * Анализ выбросов и удаление их при необходимости.

        * кол-во сделанных заказов;
        * общая сумма по всем заказам;
        * количество дней от даты последнего заказа.
* Определение категорий товаров последнего заказа для каждого покупателя (для формирования персональных предложений).
* Исследовательский анализ данных:
    * Анализ описательной статистики числовых показателей профилей покупателей.
    * Категоризация пользователей по каждому из показателей:
        * по кол-ву заказов (мало, средне, много);

        * по давности последней покупки (на днях, сравнительно недавно, давно).

* Статистический анализ.

### 2.5. Категоризация товаров

Для формирования персональных предложений для покупателей товары нужно разбить по категориям.

Сначала переведем названия товаров в нижний регистр.

```python
data['product'] = data['product'].str.lower()
```

Выделим из названий товаров первые два слова и проанализируем их.

```python
(data['product'].str.split(' ',expand=True)[0]+' '+data['product'].str.split(' ',expand=True)[1]).sort_values(ascending=True).unique()
```

Есть одни и те же слоыа, записанные по-разному. Нужно заменить:
* 'ё' на 'е';
* 'наматрицник-чехол' - "наматрасник"; 
* 'tepмокружка' - 'термокружка'

```python
data['product'] = data['product'].str.replace('ё','е')
data['product'] = data['product'].str.replace('наматрацник','наматрасник')
data['product'] = data['product'].str.replace('наматрицник-чехол','наматрасник')
data['product'] = data['product'].str.replace('tepмокружка','термокружка')
```

Создадим колонку product_category для категории товара, поместим туда значение 0. 

Напишем функции:
* set_category(words,category) - присваивает категорию, указанную в параметре category, если в названии есть слова из списка words.
* correct(words,category) - функция для корректировки неправильно присвоенной категории.

```python
data['product_category']=0

def set_category(words,category):
    for word in words:
        data.loc[(data['product'].str.find(word)>-1) & (data['product_category']==0), 'product_category'] = category     
        
def correct(words,category):
    for word in words:
        data.loc[data['product'].str.find(word)>-1, 'product_category'] = category     
```

### 2.6. Проверка данных на аномалии и выбросы

Сначала создадим pivot таблицу, где колонками будут номера категорий. Для каждого покупателя будут 1 в тех колонках, категории которых есть в таблице data_group_customer (в остальных колонках для этих покупателей будут 0).

```python
pivot = df.pivot_table(index = 'customer_id', columns = 'product_category', values='date', aggfunc='count').fillna(0).reset_index()
pivot.head()
```

Дадим колонкам строковые названия, чтобы можно было работать с ними.

```python
pivot.columns = ['customer_id','a1','a2','a3','a4','a5','a6','a7','a8']
pivot.head()
```

Теперь сделаем в pivot следующие замены:
* 0.0 заменим на пустую строку;
* 1.0 заменим на номер категории (совпадает с номером в названии колонки: а1 - 1).


```python
columns = ['a1','a2','a3','a4','a5','a6','a7','a8']
for i,column in enumerate(columns):   
    pivot[column] = pivot[column].astype('int')     # приводим к типу int, чтобы в колонках было число без точки (0 вместо 0.0)          
    pivot.loc[pivot[column]!=0,column] = i+1        # прибавляем 1, т.к. нумерация у i с 0
    pivot[column] = pivot[column].astype('str')     # приводим колонку к строковому типу для последующей конкатенации в список
    pivot.loc[pivot[column]=='0',column] = ''       # в колонки с 0 записываем пустую строку
pivot.head()
```

Теперь получим список категорий, сконкатенировав значения в колонках.

```python
pivot['actual_categories'] = ''
for column in columns:
    pivot['actual_categories'] += pivot[column]
pivot.head()
```

Для каждого покупателя категории образуют нужный список.

### 4.4. Выделение сегментов пользователей

Для формирования сегментов на основе истории покупок создадим категории покупателей по следующим признакам:
* давность последней покупки;
* количество покупок;
* общая сумма покупок.

Категории будут такими:
* 1 - лучшие показатели;
* 2 - средние показатели;
* 3 - худшие показатели.

Уровень категории показателей будем определять так:
* Давность покупки разделим по сегментам исходя из 25 и 50 перцентилей:
    * до 25 - 1 категория;
    * от 25 до 50 - 2 категория;
    * от 50 - 3 категория.
* Общую сумму разделим по сегментом по 50 и 75 перцентилям:
    * от 75 - 1 категория (большие суммы);
    * от 50 до 75 - 2 категория (средние суммы);
    * до 50 - 3 категория (маленькие суммы).
* Количество покупок раделим на две категории, т.к. у большинства покупателей этот показатель = 1:
    * кол-во покупок =1  -  1 категория;
    * кол-во покупок >1  -  2 категория.


Напишем функцию set_category(column), которая присвоит категорию заданной колонке column. Значения перцентилей будут в списке limits, значения категорий - в списке categories.

```python
def set_category(column):
    if column <= limits[0]:
        return categories[0]
    if (column > limits[0]) & (column <= limits[1]):
        return categories[1]
    if column > limits[1]:
        return categories[2]     
```

Определим категории для давности покупок.

```python
limits = np.percentile(data_group_customer['days_ago'], [25, 50])
print('Перцентили [25, 50] для days_ago:',limits)
categories = [1,2,3]

data_group_customer['days_ago_category'] = data_group_customer['days_ago'].apply(set_category)
data_group_customer.head()
```

```python
limits = np.percentile(data_group_customer['total_cost'], [50, 75])
print('Перцентили [50, 75] для total_cost:',limits)
categories = [3,2,1]

data_group_customer['total_cost_category'] = data_group_customer['total_cost'].apply(set_category)
data_group_customer.head()
```

```python
data_group_customer.loc[data_group_customer['orders'] > 1, 'orders_category'] = 1
data_group_customer.loc[data_group_customer['orders'] == 1, 'orders_category'] = 2
data_group_customer.head()
```

Теперь для каждого пользователя сформируем сегмент - строку из трех цифр, где:
* первая цифра - категория давности последнего посещения (days_ago_category);
* вторая цифра - категория общей суммы покупок (total_cost_category);
* третья цифра - категория количества заказов (orders_category).

```python
data_group_customer['segment'] = data_group_customer['days_ago_category'].astype('str') + \
                                 data_group_customer['total_cost_category'].astype('str') +\
                                 data_group_customer['orders_category'].astype('int').astype('str')
data_group_customer.head()
```

## 5. Статистический анализ данных

Проверим гипотезу о том, что выручка от покупателей прибыльных сегментов больше выручки от остальных покупателей.

Выделим данные в отдельные датафреймы.

```python
# покупатели прибыльных сегментов
profitable = data_group_customer[data_group_customer['segment'].isin(profitable_segments)]
profitable.head()
```

```python
# покупатели остальных сегментов
other = data_group_customer[~data_group_customer['segment'].isin(profitable_segments)]
other.head()
```

```python
profitable['avg_check'].describe()
```

```python
other['avg_check'].describe().T
```

```python
# функция для вывода гистограмм в две колонки
# dfs - список датафреймов, для которых строим гистограммы
# column - колонка, которую анализируем
# titles - список заголовков для гистограмм
def hists(dfs,column,titles):
    sns.set_style('darkgrid')  
    sns.set_palette('coolwarm')
    fig, axes = plt.subplots(1, 2, figsize=(8,4),dpi=144)  
    axe = axes.ravel()
    plt.yticks(fontsize=6)
    for i,df in enumerate(dfs):

        ax = df[column].plot( kind = 'hist',  bins = 50, alpha = 0.5, ax = axe[i], grid = True)
        
        ax.set_ylabel('частота',fontsize=8)
        ax.set_xlabel('прибыль, руб.',fontsize=8)
        ax.tick_params(labelsize=7)
        ax.set_title(titles[i],fontsize=10)

    plt.subplots_adjust(wspace=0.3)
```

Построим гистограммы распределения суммы покупок для двух групп покупателей.

```python
hists([profitable,other],'total_cost',['Прибыльные покупатели','Остальные покупатели'])
```

Из-за выбросов в данных графики не совсем понятны, ограничим значения прибыли:
* для прибыльных покупателей - до 10000;
* для остальных - до 4000.

```python
hists([profitable.query('total_cost < 10000'),other.query('total_cost < 4000')],'total_cost',['Прибыльные покупатели','Остальные покупатели'])
```

На графике распределения не являются нормальными.

Проверим распределения на нормальность с помощиью теста Шапиро-Уилка. Уровень значимости выберем 0.05.

```python
def shapiro_test(value,text):
    result = st.shapiro(value)
    print(f'p-value для {text}: {result.pvalue:.10f}')
    if result.pvalue < alpha:
        print('Распределение не является нормальным')
    else: print('Распределение близко к нормальному')

alpha = 0.05
    
shapiro_test(profitable['avg_check'],'profitable')
print()
shapiro_test(other['avg_check'],'other')
```

Т.к. распределения не являются нормальными, проверим данные на равенство дисперсий с помощью теста Левене, уровнь значимости возьмем 0.05.

```python
result = st.levene(profitable['avg_check'], other['avg_check'], center='mean', proportiontocut=0.05)
print(f'p_value = {result.pvalue:.10f}')
if result.pvalue < alpha:
    print('Дисперсии выборок не равны')
else: print('Дисперсии выборок равны')
```

Теперь проверим основную гипотезу. Так как дисперсии выборок не равны, зададим параметр equal_var = False

Нулевая гипотеза формулируется исходя из равенства значений, поэтому гипотезы будут такими:

<b>Н0: Выручка от покупателей прибыльных сегментов и выручка от остальных покупателей не отличаются.</b>

<b>Н1: Выручка от покупателей прибыльных сегментов больше выручки от остальных покупателей.</b>

```python
results = st.ttest_ind(profitable['avg_check'],other['avg_check'],equal_var = False)
p_value = results.pvalue
if p_value < alpha/2:
    print(f"p_value = {p_value:.10f}: Недостаточно оснований для принятия нулевой гипотезы. Принимаем альтернативную.")
else:
    print(f"p_value = {p_value:.10f}: Недостаточно оснований для того, чтобы отвергнуть нулевую гипотезу.")
```

Покупатели с высоким средним чеком приносят основную часть прибыли магазина независимо от количества заказов и давности посещения.

```python
print(f"Средний чек прибыльных покупателей составляет {profitable['avg_check'].mean():.2f}")
```

Можно выделить группу перспективных клиентов, которые делали заказ недавно (не более 3 месяцев назад) или не слишком давно (не более 7 месяцев назад) на большие суммы (около 4106.34 руб.). 

В целом наибольшим спросом пользуются растения и хозтовары (это товарные категории с большим ассортиментом).

Среди прибыльных покупателей наибольшим спросом пользуются хозтовары (2 место по количеству ассортимента).

<b>Рекомендации:</b>
* Покупателям прибыльных и перспективных сегментов можно делать персональные предложения:
    * сделать рассылку с предложением скидок на товары актуальной для покупателя категории;
    * при расширении ассортимента товаров какой-либо категории присылать уведомление об этом заинтересованнм покупателям.
* Можно увеличить ассортимент хозтоваров, т.к. они пользуются спросом у прибыльных и перспективных клиентов.
