# Проект "Компьютерные игры"
**Цель проекта:** Выявление потенциально популярного продукта среди компьютерных игр.

На основании данных из открытых источников и исторических данных о продажах игр, оценки пользователей и экспертов, жанры и платформы нужно выявить закономерности, определяющие успешность игры. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Задачи проекта:**
* Исследование жизненного цикла игр;
* Анализ популярных платформ и жанров игр;
* Исследование влияние на продажи отзывов пользователей и критиков;
* Определение топ-5 популярных жанров и платформ.

Данные представлены до конца декабря 2016 года. Нужно отработать принцип работы с данными, спрогнозировать продажи на 2017 год по данным 2016-го.

<b>Описание данных<b/>
* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

<b>Пояснения к данным:</b> 
* В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».
* Данные за 2016 год могут быть неполными.

### 3. Исследовательский анализ данных

#### 3.1. Исследование количества выпускаемых игр по годам

Посмотрим, сколько игр выпускалось в разные годы. Создадим датафрейм years_games_count, который содержит количество выпущенных игр для каждого года.

```python
years_games_count = data.groupby(['year_of_release','name'])['platform'].count().reset_index()
years_games_count.tail()
```

```python
years_games_count = years_games_count.groupby('year_of_release')['name'].count().reset_index()
years_games_count.columns = ['year_of_release','games_count']
years_games_count.tail()
```

#### 3.4. Исследование зависимости продаж от отзывов пользователей и критиков

Напишем функцию corr_func, которая считает корреляцию между объемом продаж и отзывами критиков/пользователей, а также выводит диаграмму рассеяния для этих параметров.

```python
def corr_func(platform):
    #str_query_critic - строка для запроса метода query() для нужной платформы для отзывов критиков, которые известны
    str_query_critic = 'platform =="'+platform+'" and critic_score<999 and year_of_release>=2015'
    #str_query_user - строка для запроса метода query() для нужной платформы для отзывов пользователей, которые известны
    str_query_user = 'platform =="'+platform+'" and user_score<99 and year_of_release>=2015'
    
    sns.set(font_scale=1.5) 
    fig, axes = plt.subplots(1, 2, figsize=(12,6))
    axe = axes.ravel()

    sns.scatterplot(data=data.query(str_query_critic),x='critic_score',y='total_sales',ax=axe[0])
    axe[0].set_xlabel('Отзывы критиков',fontsize=15)
    axe[0].set_ylabel('Объем продаж (млн копий)',fontsize=15)
    sns.scatterplot(data=data.query(str_query_user),x='user_score',y='total_sales',ax=axe[1]) 
    axe[1].set_xlabel('Отзывы пользователей',fontsize=15)
    axe[1].set_ylabel('Объем продаж (млн копий)',fontsize=15)
    
    corr = data.query(str_query_critic)['total_sales'].corr(data.query(str_query_critic)['critic_score'])
    print('Корреляция между отзывами критиков и продажами:      ',corr)   
    corr = data.query(str_query_user)['total_sales'].corr(data.query(str_query_user)['user_score'])
    print('Корреляция между отзывами пользователей и продажами:',corr)
```

### 4. Портрет пользователя каждого региона

Для наглядного сравнения долей напишем функцию regions, которая
* для каждого региона выведет топ-5 популярных платформ/жанров/ESRB-рейтингов и продажи по ним;
* построит круговые диаграммы для долей продаж по каждой платформе/жанру/ESRB-рейтингу из топ-5 для каждого региона. 

Входные параметры функции:
* df - исследуемый датафрейм;
* parameter - столбец, долю которого исследуем ('platform','genre' или 'rating');
* title - название параметра для отображения в заголовке диаграммы;

```python
def regions(df,parameter,title):
    num=5 # считаем для топ-5
    regs = ['na','eu','jp'] # список названий регионов
    
    all_regions = pd.DataFrame(index=range(0,num)) #Таблица для данных по продажам по всем регионам
    fig, axes = plt.subplots(1, 3, figsize=(8,6),dpi=144)  
    
    df = df[df['year_of_release']>=2015]
    
    for i,reg in enumerate(regs):
        #region - таблица по продажам текущего региона
        region = df.groupby([parameter])[reg+'_sales'].agg('sum').reset_index() \
                   .sort_values(by=reg+'_sales', ascending = False).head(num)
        region.columns = [reg+'_'+parameter,reg+'_sales']
        region.index = list(range(0, num))
        all_regions = all_regions.join(region)
        
        colors = sns.color_palette('pastel')[0:num]
        labels = region[reg+'_'+parameter]       
        plt.rcParams['font.size']=8
        ax = axes[i]
        ax.pie(region[reg+'_sales'], labels=labels, colors=colors, autopct='%.0f%%', textprops={'fontsize': 8})      
        ax.set_title(reg.upper()+' ('+title+', %)',fontsize=10)
        
    display(all_regions)
```

### 5. Проверка гипотез

Нужно проверить две гипотезы:
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


#### 5.1. Проверка гипотезы "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

Нужно проверить гипотезу "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

Так как нулевая гипотеза формулируется исходя из равенства значений, то формулировку оставляем:

<b>Н0: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".</b>

Альтернативная гипотеза будет такой:

<b>Н1: "Средние пользовательские рейтинги платформ Xbox One и PC различаются".</b>

Для проверки гипотезы выведем основные метрики.

```python
data[data['platform'].isin(['XOne','PC'])&(data['user_score']<99)&(data['year_of_release']>=2015)]. \
     pivot_table(index= 'platform', values= 'user_score', aggfunc= ['count', 'mean', 'var','std', 'median'])
```

По метрикам можно сделать вывод:
* Размеры выборок сильно отличаются.
* Различия в средних значениях, дисперсиях незначительно отличаются.

Создадим датафреймы для каждой исследуемой платформы.

```python
PC = data[(data['platform']=='PC')&(data['user_score']<99)&(data['year_of_release']>=2015)]
XOne = data[(data['platform']=='XOne')&(data['user_score']<99)&(data['year_of_release']>=2015)]
```

Построим распределение пользовательских рейтингов для каждой платформы.

```python
plt.figure(1, figsize=(16,8))
ax1 = PC['user_score'].plot( kind = 'hist',  bins = 50, alpha = 0.5, label='PC', legend=True, grid = True)
XOne['user_score'].plot( kind = 'hist', ax=ax1, bins = 50, alpha = 0.5, label='XOne', legend =True, grid = True, \
                         xlabel='user_score')
ax1.set_title('Отзывы пользователей по платформам PC и XOne')
ax1.set_xlabel('Отзывы пользователей')
```

На графике у платформы XOne распределение со скосом влево, т.е. не является правильным нормальным распределением.
У платформы PC график сильно отличается от нормального распределения.

Проверим распределения на нормальность с помощью теста Шапиро-Уилка. Уровень значимости возьмем alpha = 0.5.

```python
result_PC = shapiro(PC['user_score'])
result_XOne = shapiro(XOne['user_score'])
print('p_value для   PC:',result_PC.pvalue)
print('p_value для XOne:',result_XOne.pvalue)
```

В обоих случаях p_value < alpha. Значит, распределение не является нормальным. Cравним дисперсии выборок с помощью теста Левене, уровнь значимости возьмем 0.05.

```python
result = st.levene(PC['user_score'], XOne['user_score'], center='mean', proportiontocut=0.05)
print('p_value = ', round(result.pvalue,5))
```

С помощью теста Левене проверяется нулевая гипотеза, которая говорит о том, что дисперсии выборок равны. Мы получили p_value = 0.01361 (< 0.05). Значит, по тесту Левене дисперсии выборок не равны.

Создадим функцию для проверки гипотез.

```python
# df1,df2 - проверяемые датафреймы
# is_equal_var - True, если equal_var=True
#                False, если equal_var=False
def hipothesis(df1,df2,is_equal_var):
    if is_equal_var:
        equal_var = True
    else:
        equal_var = False
    results = st.ttest_ind(df1,df2,equal_var = equal_var)
    alpha = 0.05
    p_value = results.pvalue
    if p_value < alpha:
        print("p_value =",p_value,": Недостаточно оснований для принятия нулевой гипотезы. Принимаем альтернативную.")
    else:
        print("p_value =",p_value,": Недостаточно оснований для того, чтобы отвергнуть нулевую гипотезу.") 
```

Поскольку размеры выборок сильно отличаются и дисперсии тоже не равны, проверку гипотезы будем проводить с параметром equal_var=False.

```python
hipothesis(PC['user_score'],XOne['user_score'],False)
```

Вывод: средние пользовательские рейтинги платформ Xbox One и PC одинаковы.

#### 5.2. Проверка гипотезы "Средние пользовательские рейтинги жанров Action и Sports разные".

Нужно проверить гипотезу "Средние пользовательские рейтинги жанров Action и Sports разные".

Так как нулевая гипотеза формулируется исходя из равенства значений, то сформулируем ее так:

<b>Н0: "Средние пользовательские рейтинги жанров Action и Sports равны".</b>

Альтернативная гипотеза будет такой:

* дисперсии отличаются незначительно.

Создадим датафреймы для каждоого исследуемого жанра.

```python
Action = data[(data['genre']=='Action')&(data['user_score']<99)&(data['year_of_release']>=2015)]
Sports = data[(data['genre']=='Sports')&(data['user_score']<99)&(data['year_of_release']>=2015)]
```

Построим распределение пользовательских рейтингов для каждой платформы.

```python
plt.figure(1, figsize=(16,8))
ax1 = Action['user_score'].plot( kind = 'hist', title = 'Action', bins = 50, alpha = 0.5, label='Action', legend=True, grid = True)
Sports['user_score'].plot( kind = 'hist', title = 'Sports', ax=ax1, bins = 50, alpha = 0.5, label='Sports', \
                           legend =True, grid = True)
ax1.set_title('Отзывы пользователей по жанрам Action и Sports')
ax1.set_xlabel('Отзывы пользователей')
```

На графике распределение со скосом влево у данных обеих платформ, т.е. не является правильным нормальным распределением.

Проверим распределения на нормальность с помощью теста Шапиро-Уилка. Уровень значимости возьмем alpha = 0.5.

```python
result_Action = shapiro(Action['user_score'])
result_Sports = shapiro(Sports['user_score'])
print('p_value для Action:',result_Action.pvalue)
print('p_value для Sports:',result_Sports.pvalue)
```

В обоих случаях p_value < alpha. Значит, распределение не является нормальным, и нужно сравнить дисперсии выборок с помощью теста Левене. Вызовем метод levene() с уровнем значимости 0.05.

```python
result = st.levene(Action['user_score'], Sports['user_score'], center='mean', proportiontocut=0.05)
print('p_value = ', result.pvalue)
```

С помощью теста Левене проверяется нулевая гипотеза, которая говорит о том, что дисперсии выборок равны. Мы получили p_value гораздо меньше 0.05. Значит, по тесту Левене дисперсии выборок не равны, поэтому при проверке гипотезы зададим параметр equal_var=False.
